```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow_data_validation as tfdv
```

```python
# Load data from local directory:
df = pd.read_csv("../data/raw_data/raw_census.csv", skipinitialspace=True)
df.head()
```

```python
# Use TFDV to look at the statistics of the dataset:
stats = tfdv.generate_statistics_from_dataframe(dataframe=df, n_jobs=-1)
tfdv.visualize_statistics(stats)
```

```python
# Use TFDV to look at the schema of the dataset:
schema = tfdv.infer_schema(statistics=stats)
tfdv.display_schema(schema=schema)
```

```python
# Use TFDV to check anomalies:
anomalies = tfdv.validate_statistics(statistics=stats, schema=schema)
tfdv.display_anomalies(anomalies=anomalies)
```

#### Data Preprocessing

```python
# Drop rows containing "?" values:
pre_shape = df.shape
df.replace(to_replace="?", value=np.nan, inplace=True)
df.dropna(axis=0, inplace=True)
post_shape = df.shape

print(f"{pre_shape[0] - post_shape[0]} rows has been dropped.")
```

```python
# Drop columns containing too many zero values:
df.drop(columns=["capital-loss", "capital-gain"], inplace=True)

# Drop column "fnlgt" which seems like an ID column
df.drop(columns=["fnlgt"], inplace=True)
```
